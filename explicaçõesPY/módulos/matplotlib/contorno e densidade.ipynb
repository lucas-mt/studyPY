## visualizando funções tridimensionais em plots bidimensionais
---

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

para visualizar funções tridimensionais através de curvas, é possível usar a função `.contour()`, que recebe três valores: um conjunto de valores em x, um conjunto de valores em y e um conjunto de valores em z. x e y representam as posições no gráfico e z representa a altura dos níveis.

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
z = f(X, Y)

plt.contour(X, Y, z, colors='black')
```

é necessário usar a função de numpy `.meshgrid()`, que constrói conjuntos de valores bidimensionais a partir de matrizes unidimensionais, para que z tenha a quantidade exata de valores, podendo, assim, construir o gráfico.

observe que foi usado o parâmetro `colors=`, que unifica a cor de todas as linhas do gráfico. Desta forma, valores negativos são representados por pontilhados; já valores positivos, por linhas. Se nã usar este parâmetro, o plot é colorido:

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
z = f(X, Y)

plt.contour(X, Y, z)
```

pode, ainda, usar o parâmetro `cmap=`, que recebe duas cores e, assim, o gráfico se mostrará em um degradê. Este tipo de visualização é interessante em gráficos com valores centralizados.

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
z = f(X, Y)

plt.contour(X, Y, z, cmap='RdGy')
```

para saber todas as combinações de cores, abra o `IPython`, digite `plt.cm.` e aperte TAB.

a função `.contourfl()` funciona da mesma forma, mas preenche os espaços vazios do gráfico:

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
z = f(X, Y)

plt.contourf(X, Y, z)
plt.colorbar()
```

neste exemplo, foi adicionado o `.colorbar()` que ajuda a entender os valores pela cor do gráfico. Neste caso, quanto mais verde, maior o valor, quando mais azul, menor. Para mais informações sobre esta função, olhe o jupyter notebook scatter.ipynb.

pode, ainda, usar a função `.imshow()` que apresenta o gráfico como uma foto. Neste caso, não recebe x ou y, apenas os valores de z. Logo, é necessário passar manualmente os limites de x e y, neste caso, de 0 a 5 cada, através do parâmetro `extent=`. É, também, necessário usar o parâmetro `origin=` para corrigir a definição padrão de matriz de imagens, onde a origem está no canto superior esquerdo, e não no canto inferior esquerdo, como na maioria dos gráficos de contorno. Isso deve ser alterado ao mostrar dados em grade.

```python
plt.imshow(z, extent=[0, 5, 0, 5], cmap='RdGy')
plt.colorbar()
```
