## matplotlib.pyplot: fill
---

```python
import matplotlib.pyplot as plt
```

usando os métodos `fill` é possível preencher espaços entre plots, área embaixo das linhas, etc.

#### fill_between()
---

preenche o espaço que fique entre os parâmetros passados a `fill_between()`. Por exemplo:

```python
x = [i for i in range(-5, 6)]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='red')
plt.fill_between(x, y1, y2)
plt.grid()
plt.show()
```

neste exemplo, toda a área entre x² e x³ está sendo preenchido em relação à abscissa x.

pode ser acrescentado uma transparência no preenchimento usando `alpha=` em `fill.between()`, podendo ser usado caso o preenchimento da área cubra alguma informação importante.

```python
x = [i for i in range(-5, 6)]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='red')
plt.fill_between(x, y1, y2, alpha=0.2)
plt.grid()
plt.show()
```

`fill_between` também aceita argumentos como `label=` - sempre lembrando de incluir `plt.legend()` - e `color(s)=`, por exemplo.
