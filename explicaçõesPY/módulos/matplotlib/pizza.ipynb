## matplotlib.pyplot: pizza
---

```python
import matplotlib.pyplot as plt
```

para mostrar um gráfico no formato de pizza basta usar `plt.pie()` ao invés de `plt.plot()`

```python
partes = [60, 40]

plt.pie(partes)
plt.show()
```

como o gráfico está sem nenhuma identificação, dentro de `plt.pie()` pode-se acrescentar `labels=<lista>` que recebe uma lista contendo a descrição de cada fatia:

```python
partes = [60, 40]
descrição = ['vontade', 'ação']

plt.pie(partes, labels=descrição)
plt.show()
```

se quiser fazer uma separação entre as cores use o comando `wedgeprops={'edgecolor': '<cor>'}` dentro de `plt.pie()`

```python
import matplotlib.pyplot as plt

partes = [60, 40]
descrição = ['vontade', 'ação']

plt.pie(partes, labels=descrição, wedgeprops={'edgecolor': 'black'})
plt.show()
```

`edgecolor` aceita cores como: white, blue, red, pink, magenta, brown, yellow,, grey, green, etc.

para pintar as fatias com as cores desejadas é necessário passar uma lista contendo as cores e passá-la como parâmetro para `colors=` em `plt.pie()`

```python
fatias = [60, 40, 10, 30]
descrição = ['60', '40', '10', '30']
cores = ['blue', 'green', 'magenta', 'brown']

plt.pie(fatias, labels=descrição, colors=cores)
plt.show()
```

o comando `explode=<lista>` recebe uma lista como valores que indicão se é desejado destacar uma fatia em específico. O valor padrão é 0 (zero).

```python
partes = [60, 40]
exp = [0, 0.1]

plt.pie(partes, explode = exp, wedgeprops={'edgecolor': 'black'})
plt.show()
```

é bom colocar valores pequenos, pois valores grandes podem fazer com esta fatia saia do enquadramento do gráfico:

```python
partes = [60, 40]
exp = [0, 2]

plt.pie(partes, explode = exp, wedgeprops={'edgecolor': 'black'})
plt.show()
```

o comando `shadow=`, quanndo acrescentado em `plt.pie()`, dá uma sombra no gráfico, dando a impressão de ser 3D

```python
partes = [60, 40]
exp = [0, 0.05]

plt.pie(partes, explode = exp, wedgeprops={'edgecolor': 'black'}, shadow=True)
plt.show()
```

o ângulo em que o círculo é apresentado pode ser mudado usando o comando `startangle=` em `plt.pie()`, onde deve ser passado um valor em graus

```python
partes = [60, 40]
exp = [0, 0.05]

plt.pie(partes, explode = exp, wedgeprops={'edgecolor': 'black'}, startangle=90)
plt.show()
```

para mostrar os valores em porcentagem no gráfico, o comando `autopct='%1.1f%%'` é usado em `plt.pie()`

```python
partes = [60, 40]
exp = [0, 0.1]

plt.pie(partes, explode = exp, wedgeprops={'edgecolor': 'black'}, shadow=True, autopct='%1.1f%%')
plt.show()
```
