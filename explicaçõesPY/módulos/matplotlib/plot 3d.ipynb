## gráficos 3d
---

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

para plotar gráficos 3D, deve-se:

```python
from mpl_toolkits import mplot3d
```

só então, é necessário passar o argumento `projection=3d` para qualquer objeto `.axes()`:

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

para apresentações de gráficos 3D usando jupyter, é melhor usar `%matplotlib notebook` do que `%matplotlib inline`, pois permite interagir com o gráfico.

assim, para apresentar gráficos em linhas, em analogia à função `.plot()`, usa-se `.plot3d()`. Da mesma forma, em analogia à função `.scatter()`, para gráficos do tipo scatter usa-se `.scatter3d()`:

```python
ax = plt.axes(projection='3d')

# os valores para o gráfico de linha
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# os valores para o gráfico scatter
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

para modificar como os valores são apresentados no gráfico, usa-se os mesmos parâmetros vistos nos jupyter notebooks scatter.ipynb e linhas.ipynb.

gráficos de contorno 3D também é bastante similiar ao que se vê no jupyter notebook contorno e densidade.ipynb, usando, aqui, a função `.contour3d`:

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

o método `view_init(<x-y_ângulo>, <z_ângulo>)` recebe dois ângulos, um em relação a x e y, e o outro em relação a z, que otimiza a forma como é visualizado o gráfico assim que a janela abre:

```python
ax.view_init(60, 35)
fig
```

```python
ax.view_init(45, 45)
fig
```

pode-se, ainda, usar a função `.plot_wireframe()`, que recebe os valores de x, y e z, assim como as outras funções, e plota um gráfico com sua superfície em grade:

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

ou, então, usar a função `.plot_surface()`, que funciona igual `.plot_wireframe()`, mas que preenche a superfície do gráfico com polígonos:

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

quando os valores não tão bem distribuídos, usar `.plot_surface()` pode ser um problema por criar quadriláteros. Para resolver isto, pode-se usar a função `.plot_trisurf()`, que toma três pontos dos valores passados e cria triângulos que preenchem a superfície do gráfico:

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)

ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis')
```

desta forma:

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)

ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, cmap='viridis')
```
