## colorbar
---

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

colorbar são importantes, pois quando um gráfico tem suas informações baseadas em cores, este pode mostrar o que cada cor significa. o `.colorbar()` é uma extensão de `.axes()`, por isso, aquele tem as mesmas funcionalidade deste.

A forma mais simples de aplicar o colorbar é através da função `.colorbar()`:

```python
from random import randint
import pandas as pd

x = np.linspace(0, 10, 100)
y = [randint(1, 1000) for _ in range(100)]
cor_tam = pd.Series([randint(1, 9) for _ in range(100)])

plt.scatter(x, y, c=cor_tam, s=(50*cor_tam))
plt.colorbar()
```

pode usar o parâmetro `cmap=`, que mapeia as cores para aplicá-las no gráfico, sendo necessário passar as cores em que deseja o degradê. Mas informações sobre no jupyter notebook contornos e densidade.ipynb.

```python
x = np.linspace(0, 10, 100)
y = [randint(1, 1000) for _ in range(100)]
cor_tam = pd.Series([randint(1, 9) for _ in range(100)])

plt.scatter(x, y, c=cor_tam, cmap='YlGnBu', s=(50*cor_tam))
plt.colorbar()
```

```python
x = np.linspace(0, 10, 100)
y = [randint(1, 1000) for _ in range(100)]
cor_tam = pd.Series([randint(1, 9) for _ in range(100)])

plt.scatter(x, y, c=cor_tam, cmap='jet', s=(50*cor_tam))
plt.colorbar()
```

```python
x = np.linspace(0, 10, 100)
y = [randint(1, 1000) for _ in range(100)]
cor_tam = pd.Series([randint(1, 9) for _ in range(100)])

plt.scatter(x, y, c=cor_tam, cmap='viridis', s=(50*cor_tam))
plt.colorbar()
```

`viridis` é o modelo padrão de `cmap`, como ser visto comparando este plot com o primeiro.

É possível tornar as pontas do colorbar em setas através do parâmetro `extend=`:

```python
x = np.linspace(0, 10, 100)
y = [randint(1, 1000) for _ in range(100)]
cor_tam = pd.Series([randint(1, 9) for _ in range(100)])

plt.scatter(x, y, c=cor_tam, s=(50*cor_tam))
plt.colorbar(extend='both')
```

observe que em todos os colorbar até agora as cores se apresentam em um degradê. Se desejar apresentar cada tom explicitamente, deve-se usar como argumento `plt.cm.get_cmap(<cores>, <divisões>)` para o parâmetro `cmap=`:

```python
x = np.linspace(0, 10, 100)
y = [randint(1, 1000) for _ in range(100)]
cor_tam = pd.Series([randint(1, 9) for _ in range(100)])

plt.scatter(x, y, c=cor_tam, cmap=plt.cm.get_cmap('Blues', 9), s=(50*cor_tam))
plt.colorbar()
```

observe que é necessário passar a cor e a quantidade de divisões que o colorbar deve ter
