## matplotlib.pyplot: Scatter plot
---

```python
from matplotlib import pyplot as plt
```

são gráficos essenciais para dados aleatórios, para isso basta substituir `.plot()` por `.scatter()`

```python
from random import randint

x = [randint(1, 9) for _ in range(15)]
y = [randint(1, 9) for _ in range(15)]

plt.scatter(x, y)
plt.show()
```

o tamanho dos pontos pode ser alterando usando o argumento `s=` dentro de scatter, bastaando passar um valor

```python
x = [randint(1, 9) for _ in range(15)]
y = [randint(1, 9) for _ in range(15)]

plt.scatter(x, y, s=100)
plt.show()
```

o argumento `s=` pode receber uma lista com tamanhos diferentes, contanto que esta lista tenha o mesmo tamanho das listas `x` e `y`, podendo, assim, personalizar o tamanho de cada marcador:

```python
x = [randint(1, 9) for _ in range(15)]
y = [randint(1, 9) for _ in range(15)]
size = [randint(199, 999) for _ in range(15)]

plt.scatter(x, y, s=size)

plt.show()
```

para mudar a cor dos pontos, basta passar o argumento `c=` e o nome da cor:

```python
x = [randint(1, 9) for _ in range(15)]
y = [randint(1, 9) for _ in range(15)]

plt.scatter(x, y, c='green')
plt.show()
```

o argumento `c=` pode receber uma lista de cores, contanto que essa lista seja do mesmo tamanho de das listas `x` e `y`, permitindo que cada ponto tenha uma cor única:

```python
x = [randint(1, 9) for _ in range(15)]
y = [randint(1, 9) for _ in range(15)]
colors = [randint(1, 9) for _ in range(15)]

plt.scatter(x, y, c=colors)
plt.show()
```

abaixo do comando de plot pode ser acrecentado um comando para mostrar o que cada cor significa, por exemplo, satisfação:

```python
x = [randint(1, 9) for _ in range(15)]
y = [randint(1, 9) for _ in range(15)]
colors = [randint(1, 9) for _ in range(15)]

plt.scatter(x, y, c=colors)
cbar = plt.colorbar()
cbar.set_label('satisfaction')

plt.show()
```

para mudar o estilo dos pontos pode ser usado o argumento `marker=` e os formatos possíveis podem ser vistos no notebook *linhas.ipynb*

```python
x = [randint(1, 9) for _ in range(15)]
y = [randint(1, 9) for _ in range(15)]

plt.scatter(x, y, marker='+')
plt.show()
```

 argumento `edgecolor=` seguido de uma cor pinta as bordas da marcação


```python
x = [randint(1, 9) for _ in range(15)]
y = [randint(1, 9) for _ in range(15)]

plt.scatter(x, y, edgecolor='magenta')
plt.show()
```

o argumento `alpha=` recebe um valor que tornará a coloraçã da marcação mais clara ou mais escura:

```python
x = [randint(1, 9) for _ in range(15)]
y = [randint(1, 9) for _ in range(15)]

plt.scatter(x, y, alpha=0.25)
plt.show()
```

#### usando diferentes escalas
---

usando os comandos `plt.xscale()` e `plt.yscale()` é possível mudar a escalar de valores casos estes sejam muito grandes ou muito pequenos:

```python
#sem mudar a escala
x = [66243315, 65278046, 89470577, 57593133, 78638814, 23025806, 54890726, 71503367, 81145945, 12118630, 57678894, 43209819, 91984050, 37599330, 97247659]
y = [24025004, 27377431, 80234761, 99849307, 50729542, 62463595, 80771235, 35566428, 91964994, 73015305, 85030441, 30731715, 91394987, 78874840, 60064544]
size = [randint(199, 999) for _ in range(15)]

plt.scatter(x, y, s=size)

plt.show()
```

assim, ficando:

```python
#com mudança de escala
x = [66243315, 65278046, 89470577, 57593133, 78638814, 23025806, 54890726, 71503367, 81145945, 12118630, 57678894, 43209819, 91984050, 37599330, 97247659]
y = [24025004, 27377431, 80234761, 99849307, 50729542, 62463595, 80771235, 35566428, 91964994, 73015305, 85030441, 30731715, 91394987, 78874840, 60064544]
size = [randint(199, 999) for _ in range(15)]

plt.scatter(x, y, s=size)
plt.xscale('log')
plt.yscale('log')

plt.show()
```

as escalas permitidas são: 'linear', 'log', 'symlog', 'asinh', 'logit', 'function', 'functionlog'.
