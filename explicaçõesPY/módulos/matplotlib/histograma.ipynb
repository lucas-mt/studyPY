## matplotlib.pyplot: histograma
---

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

usando o comando `plt.hist()` ao invés de `plt.plot()` um histograma é plotado.

usando o comando `bins=` em `hist()` é possível dividir o histograma em partes iguais, se não der para observar use `edgecolor=` passando uma cor:

```python
x = [i for i in range(18, 79, 2)]

plt.hist(x, bins=5, edgecolor='red')
plt.show()
```

basicamente, o que o histograma informa é que há 7 informações entre os valores de 66 a 78, por exemplo, e, assim, sucessivamente.

`bins=` pode receber uma lista de valores para essas comparações

```python
x = [i for i in range(18, 79, 2)]
b = [20, 30, 40, 50, 60, 70]

plt.hist(x, bins=b, edgecolor='red')
plt.show()
```

uma forma de facilitar a leitura é colocar `log=True` em `hist()`

```python
x = [i for i in range(18, 79, 2)]
b = [20, 30, 40, 50, 60, 70]

plt.hist(x, bins=b, edgecolor='red', log=True)
plt.show()
```

uma forma de marcar um ponto em específico é usando o método `plt.axvline()`

```python
x = [i for i in range(18, 79, 2)]
b = [20, 30, 40, 50, 60, 70]

plt.hist(x, bins=b, edgecolor='red')
plt.axvline(max(x), color='green', label='valor máximo')
plt.legend()
plt.show()
```

observe que, neste exemplo, o valor máximo nem está dentro das observações.

pode ser usado, também, o método `plt.axhline()` com o mesmo objetivo, porém, a linha aparecerá na horizontal.

a expessura desta linha pode ser mudada a partir de `linewidth=` em `axvline()`

```python
x = [i for i in range(18, 79, 2)]
b = [20, 30, 40, 50, 60, 70]

plt.hist(x, bins=b, edgecolor='red')
plt.axvline(max(x), color='green', label='valor máximo', linewidth=10)
plt.legend()
plt.show()
```

lembrando que o parâmetro `alpha=`, que torna o gráfico transparente, funciona aqui, seu uso é interessante quando há mais de um plot sendo mostrado:

```python
import numpy as np

x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(alpha=0.3,bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

#### histogramas bidimensionais
---

uma forma simples de fazwer histogramas bidimensionais é usando a função `hist2d()` que recebe dois valores x e y, deve receber em quandas partes deve ser dividido usando `bins=`, e qualquer outro parâmetro já visto antes: 

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

observe que o formato para cada bin é um quadrilátero. Pode-se usar a função `.hexbin()`, que funciona como a função `.hist2d()`, mas o formato dos bins fica sendo um hexágono:

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

a opção `gridsize=` recebe o tamanho de cada bin:

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hexbin(x, y, gridsize=10, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```
