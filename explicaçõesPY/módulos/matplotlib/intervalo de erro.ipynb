## barras de erro
---

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

é uma forma de mostrar possíveis erros em um valor, por exemplo 71 ± 0.025. Para isto, basta trocar `.plot()` por `.errorbar()`, que deve receber os valores para os eixos x e y, e o erro através do parâmetro `yerr=`:

```python
import numpy as np

fig, ax = plt.figure(), plt.axes()

x = [i for i in range(1, 10)]

ax.errorbar(x, np.sin(x), yerr=0.5)
```

pode, ainda, receber o parâmetro `fmt=` que aceita os códigos mostrados nos jupyter notebooks linhas.ipynb e scatter.ipybn para modificar o formato das barras de erro:

```python
fig, ax = plt.figure(), plt.axes()

x = [i for i in range(1, 10)]

ax.errorbar(x, np.sin(x), yerr=0.5, fmt='.k')
```

```python
fig, ax = plt.figure(), plt.axes()

x = [i for i in range(1, 10)]

ax.errorbar(x, np.sin(x), yerr=0.5, fmt='--^')
```

e, assim, sucessivamente.

todo tipo de modificação pode ser feito. Usando o parâmetro `color=`, pode-se especificar a cor do ponto que marca o meio da barra; já o parâmetro `ecolor=` modifica a cor da barra:

```python
fig, ax = plt.figure(), plt.axes()

x = [i for i in range(1, 10)]

ax.errorbar(x, np.sin(x), yerr=0.5, fmt='--^', color='black', ecolor='blue')
```

o parâmetro `elinewidth=` recebe um valor para afinar ou engrossar as barras:

```python
fig, ax = plt.figure(), plt.axes()

x = [i for i in range(1, 10)]

ax.errorbar(x, np.sin(x), yerr=0.5, fmt='--^', elinewidth=10)
```

usando o parâmetro `capsize=` é possível especificar o tamanho do topo e do fundo da barra:

```python
fig, ax = plt.figure(), plt.axes()

x = [i for i in range(1, 10)]

ax.errorbar(x, np.sin(x), yerr=0.5, fmt='--^', capsize=5)
```

```python
fig, ax = plt.figure(), plt.axes()

x = [i for i in range(1, 10)]

ax.errorbar(x, np.sin(x), yerr=0.5, fmt='--^', capsize=20)
```

é, também, possível fazer barras de erro horizontais através do parâmetro `xerr=`:

```python
fig, ax = plt.figure(), plt.axes()

x = [i for i in range(1, 10)]

ax.errorbar(x, np.sin(x), xerr=0.5)
```

as mesmas modificações feitas com `yerr=` são suportadas por `xerr=`:

```python
fig, ax = plt.figure(), plt.axes()

x = [i for i in range(1, 10)]

ax.errorbar(x, np.sin(x), xerr=0.5, fmt='-.D', color='green', ecolor='magenta', elinewidth=6, capsize=12)
```
