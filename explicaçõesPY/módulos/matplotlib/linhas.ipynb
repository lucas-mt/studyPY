#### .plot(x, y)
---

para plotar um gráfico 2D, basta passar como parâmetro duas listas ao método `.plot(<x>, <y>)`, onde a primeira (`<x>`) corresponde à abscissa e a segunda (`<y>`) corresponde à ordenada:

````python
lst_x = [1, 2, 3, 4, 5]
lst_y = [1, 4, 9, 16, 25]

plt.plot(lst_x, lst_y)
````

observe que é sempre importante passar o método `.show()` para que o gráfico, de fato, apareça.

em ipython, ao iniciar o código, é interessante usar `%matplotlib`, desta forma não será necessário usar o `.show()` pois o gráfico será apresentado automaticamente:
```
%matplotlib

lst_x = [1, 2, 3, 4, 5]
lst_y = [i**2 for i in lst_x]

plt.plot(lst_x, lst_y)
```
o gráfico é interativo, logo, ao ser adicionado informações na linha de comando, a plot vai sendo mudado altomaticamente.

#### legendando o gráfico
---

o método `.legend(<list>)` recebe uma lista contendo a legenda desejada que seguirá a ordem em que foi mandado plotar os gráficos

por exemplo: os valores ao quadrado aparecem primeiro no código, e os valores ao cubo aparecem por último; por isso, quando passado `.legend(['valores ao quadrado', 'valores ao cubo'])` a primeira descrição será relacionada com o primeiro gráfico. Já a segunda descrição será relacionada com o segundo gráfico:

````python
x = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
y3 = [1, 8, 27, 64, 125]

plt.plot(x, y2)
plt.plot(x, y3)
plt.legend(['valores ao quadrado', 'valores a cubo'])
````

devido ser necessário saber a ordem que os gráficos aparecem no código para que a legenda esteja correta, o melhor método para legendar gráficos é, no próprio `.plot()`, depois de apresentar as listas, incluir `lable=<legenda>`. só, então, que coloca `.legend()` mas, desta vez, sem receber parâmetros:

````python
x = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
y3 = [1, 8, 27, 64, 125]

plt.plot(x, y2, label='valores ao quadrado')
plt.plot(x, y3, label='valores ao cubo')
plt.legend()
````

`.legend()` ainda recebe alguns parâmetro, como `loc=`, para especificar onde a legenda deve aparecer:

````python
x = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
y3 = [1, 8, 27, 64, 125]

plt.plot(x, y2, label='valores ao quadrado')
plt.plot(x, y3, label='valores ao cubo')
plt.legend(loc='center')
````

pode, assim, colocar as legendas em `upper left`, `upper right`, `lower left`, `lower right`, `lower center`, `upper center`, `center left` e `center right`.

usando o parâmetro `ncol=`, pode-se dividir a legenda em colunas:

````python
x = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
y3 = [1, 8, 27, 64, 125]

plt.plot(x, y2, label='valores ao quadrado')
plt.plot(x, y3, label='valores ao cubo')
plt.legend(ncol=2)
````

se o parâmetro `fancybox=` receber o valor `True`, a caixa com as legendas fica com as bordas arredondadas. Pode, ainda, receber o parâmetro `framealpha=`, o que faz a legenda ficar transparente; o parâmetro `shadow=`, que cria uma sombra para a caixa das legendas; e o parâmetro `borderpad=` ajusta o tamanho da caixa:

````python
x = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
y3 = [1, 8, 27, 64, 125]

plt.plot(x, y2, label='valores ao quadrado')
plt.plot(x, y3, label='valores ao cubo')
plt.legend(fancybox=True, framealpha=0.5, shadow=True, borderpad=0.5)
````

passando `False` para o parâmetro `frameon=`, a caixa das legendas some e fica apenas a legenda em si. Já o parâmetro `labelspacing=` recebe valores de pixels para ajustar a distância entre cada legenda:

````python
x = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
y3 = [1, 8, 27, 64, 125]

plt.plot(x, y2, label='valores ao quadrado')
plt.plot(x, y3, label='valores ao cubo')
plt.legend(frameon=False, labelspacing=3)
````

quando há muitos plots, pode ser interessante legendar apenas alguns. Para isso, deve-se passar o comando `.plot()` para uma variável, que será passível de indexing e slicing. Desta forma, é possível escolher quais plots receberão legendas através da função `.legend()`:

````python
x = np.linspace(0, 10, 1000)
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
lines = plt.plot(x, y)

plt.legend(lines[:2], ['first', 'second'], loc='upper right');
````

observe que, pelo comando `lines[:2]`, apenas os dois primeiros plots receberam legendas, que foram passadas como uma lista de string na ordem que devem aparecer.

Outra forma de fazer isto é através da forma convencional usando o parâmetro `label=` em `.plot()`:

````python
plt.plot(x, y[:, 0], label='first')
plt.plot(x, y[:, 1], label='second')
plt.plot(x, y[:, 2:])
plt.legend(framealpha=1, frameon=True, loc='upper right');
````

se for necessário criar várias caixas de legenda, o matplotlib não permite isso diretamente, por isso, é necessário criar manualmente a segunda, terceira, quarta, etc. caixas que desejar. Para isso, basta importar a biblioteca

````python
from matplotlib.legend import Legend
````

em seguida, através do `Legend()`, cria uma variável, no exemplo a seguir `leh`, onde pode ser passado todos os parâmetros suportados pela função `.legend()`, sendo necessário passar um objeto `.axes()`, neste caso, a variável `ax`. Finalmente, deve passar a variável `leh` como parâmetro para a função `add_artist()`, que criará esta caixa de legenda extra.

````python
ax = plt.axes()

x = np.linspace(0, 10, 1000)
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
lines = plt.plot(x, y)


plt.legend(lines[:2], ['first', 'second'], loc='upper right');
leh = Legend(ax, lines[2:], ['third', 'fourth'], loc='lower right')
ax.add_artist(leh)
````

mas, não somente os estilos predeterminados pelo matplotlib podem ser usados, o usuário pode criar os próprios estilos:

````python
x = np.random.randn(1000)
plt.hist(x);

#permitindo cores.
ax = plt.axes()
ax.set_axisbelow(True)

# desenhando linhas brancas sólidas para formar a grade
plt.grid(color='w', linestyle='solid')

# escondendo os traços
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# clareando ticks e marcações
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')

# colorindo os traços e o interior do histograma
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
````

neste exemplo, a mudança no gráfico foi feito manualmente e servirá apenas para este gráfico. Se desejar criar um estilo que seja aplicável em vários gráficos, é preciso usar a função `plt.rc()`:

````python
from matplotlib import cycler

IPython_default = plt.rcParams.copy()

x = np.random.randn(1000)

colors = cycler('color',
                ['#EE6666', '#3388BB', '#9988DD',
                 '#EECC55', '#88BB44', '#FFBBBB'])
plt.rc('axes', facecolor='#E6E6E6', edgecolor='none',
       axisbelow=True, grid=True, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='gray')
plt.rc('ytick', direction='out', color='gray')
plt.rc('lines', linewidth=2)

plt.hist(x, edgecolor='#E6E6E6')
````

observe que, primeiro, foi armazenado em uma variável, neste caso, `IPython_default`, o estilo padrão do gráfico copiando a biblioteca `plt.rcParams`. Então, para cada comando `plt.rc()` foi passado o que estava sendo modificado, seguido dos parâmetros já visto anteriormente para modificar gráficos. Acrescenta-se aqui `direction=`, que recebe `in` ou `out` e mostrar se os traços dos ticks ficam para fora ou para dentro, e `linestayle=`, que recebe `solid` para o traço aparecer mais forte.

é possível armazenar os estilos criados em um arquivo .matplotlibrc

````python
#retornando ao estilo padrão
plt.rcParams.update(IPython_default)
````

#### criando objetos
---

até agora, apenas foi usado a forma mais simples de criar um gráfico, usando o próprio `plt` para isso. No entanto, é possível criar objetos que farão o mesmo serviço, tendo o bônus de simplificar na hora de modificar os gráficos. Desta forma, o gráfico acima pode ser reescrito:

````python
fig = plt.figure()
ax = plt.axes()

x = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

ax.plot(x, y2, label='valor ao quadrado')
ax.legend()
````

assim, a variável `fig` armazena o eixo, o gráfico, os textos e os rótulos; já a variável `ax` é uma caixa delimitadora com marcas de seleção e rótulos, que eventualmente conterá os elementos do gráfico que compõem nossa visualização.

apesar de `ax` funcionar da mesma forma que `plt`, algumas funções são diferentes:
plt.função()|ax.função()
---|---
xlabel|set_xlabel
ylabel|set_ylabel
xlim|set_xlim
ylim|set_ylim
title|set_title

já outros são iguais:
plt.função()|ax.função()
---|---
plot|plot
legend|legend


desta forma, pode ser melhor usar o método `.set()` com `ax`, assim, tudo isso pode ser passado de uma só vez como parâmetros:

````python
x = pd.Series(np.linspace(0, 1000, 1000))

ax = plt.axes()
ax.set(xlim=(0, 10), ylim=(-1.2, 1.2), xlabel='x', ylabel='cos(x)', title='f(x)=cos(x)')
ax.plot(x, np.cos(x))
````
