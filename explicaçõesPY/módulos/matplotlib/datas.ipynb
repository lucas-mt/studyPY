## matplotlib.pyplot: plot_date
---

```python
from datetime import datetime
import matplotlib.pyplot as plt
```

é uma forma de plotar gráficos com data, basta substituir `.plot()` por `.plot_date`

```python
x = [datetime(2019, 5, 24), datetime(2019, 5, 25), datetime(2019, 5, 26),
datetime(2019, 5, 27), datetime(2019, 5, 28), datetime(2019, 5, 29), datetime(2019, 5, 30)]
y = [1, 2, 3, 4, 5, 6, 7]

plt.plot_date(x, y)

plt.tight_layout()
plt.show()
```

nativamente, o gráfico de data é plotado usando pontos, para mudar isso pode ser usado o argumento `.linestyle=`:

```python
x = [datetime(2019, 5, 24), datetime(2019, 5, 25), datetime(2019, 5, 26),
datetime(2019, 5, 27), datetime(2019, 5, 28), datetime(2019, 5, 29), datetime(2019, 5, 30)]
y = [1, 2, 3, 4, 5, 6, 7]

plt.plot_date(x, y, linestyle='-')

plt.tight_layout()
plt.show()
```

todos os estilos de linha podem ser vistas no notebook *linhas.ipynb*

use o comando `plt.gcf().autofmt_xdate()` para corrigir a formatação das datas:

```python
x = [datetime(2019, 5, 24), datetime(2019, 5, 25), datetime(2019, 5, 26),
datetime(2019, 5, 27), datetime(2019, 5, 28), datetime(2019, 5, 29), datetime(2019, 5, 30)]
y = [1, 2, 3, 4, 5, 6, 7]

plt.plot_date(x, y)
plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()
```

para modificar a forma como a data está sendo apresentada será necessário importar um outro módulo de matplotlib:

```python
from matplotlib import dates as dt
```

o comando onde será passado o formato da data é o `dt.DateFormatter()` que deve ser passado como parâmetro os estilos de datas que podem ser vistos no notebook *datetime.ipynb*.

para que, de fato, seja mostrado no plot como é passado no comando acima, o comando `plt.gca().xaxis.set_major_formatter(<comando_acima>)`:

```python
x = [datetime(2019, 5, 24), datetime(2019, 5, 25), datetime(2019, 5, 26),
datetime(2019, 5, 27), datetime(2019, 5, 28), datetime(2019, 5, 29), datetime(2019, 5, 30)]
y = [1, 2, 3, 4, 5, 6, 7]

plt.plot_date(x, y)
plt.gcf().autofmt_xdate()
dt_fmt = dt.DateFormatter('%d %B %y')
plt.gca().xaxis.set_major_formatter(dt_fmt)

plt.tight_layout()
plt.show()
```
