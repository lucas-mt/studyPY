#### salvando o plot como uma foto
---

para salvar o plot de um gráfico como uma foto, usa-se a função `<graf>.savefig('nome.<formato>')`

```python
x = [i for i in range(1, 6)]
y = [i**2 for i in x]

fig = plt.figure()
plt.plot(x, y)

fig.savefig('meu gráfico.png')
```

observe que é necessário criar uma variável que contenha a figura para que, a partir desta variável, no exemplo, a variável `fig`, seja salvo o gráfico em formato de foto.

o formato da foto é aquele determinado no comando, e suporta os seguintes:

```python
fig.canvas.get_supported_filetypes()
```

#### limitando os valores dos eixos
---

usando `.xlim(min, max)` e `.ylim(min, max)` é possível limitar os valores que aparecem no gráfico:

```python
import numpy as np

fig, ax = plt.figure(), plt.axes()

x = [i for i in range(-1000, 1000)]
y = [np.sin(i) for i in x]

ax.plot(x, y)
```

limitando, fica:

```python
ig, ax = plt.figure(), plt.axes()

x = [i for i in range(-1000, 1000)]
y = [np.sin(i) for i in x]

plt.xlim(-10, 10)
plt.ylim(-1.5, 1.5)

ax.plot(x, y)
```

se desejar mostrar os valores em ordem decrescente, basta invester os valores: `xlim(max, lim)` e `ylim(max, lim)`:

```python
ig, ax = plt.figure(), plt.axes()

x = [i for i in range(-1000, 1000)]
y = [np.sin(i) for i in x]

plt.xlim(10, -10)
plt.ylim(1.5, -1.5)

ax.plot(x, y)
```

passando a string `tight` para `plt.axis()`, o gráfico aparece justificado:

```python
fig, ax = plt.figure(), plt.axes()

a = [i for i in range(0, 10)]

plt.axis('tight')

ax.plot(a, np.sin(a))
```
