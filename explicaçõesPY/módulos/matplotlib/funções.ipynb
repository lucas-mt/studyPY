#### salvando o plot como uma foto
---

para salvar o plot de um gráfico como uma foto, usa-se a função `<graf>.savefig('nome.<formato>')`

```python
x = [i for i in range(1, 6)]
y = [i**2 for i in x]

fig = plt.figure()
plt.plot(x, y)

fig.savefig('meu gráfico.png')
```

observe que é necessário criar uma variável que contenha a figura para que, a partir desta variável, no exemplo, a variável `fig`, seja salvo o gráfico em formato de foto.

o formato da foto é aquele determinado no comando, e suporta os seguintes:

```python
fig.canvas.get_supported_filetypes()
```

#### limitando os valores dos eixos
---

usando `.xlim(min, max)` e `.ylim(min, max)` é possível limitar os valores que aparecem no gráfico:

```python
import numpy as np

fig, ax = plt.figure(), plt.axes()

x = [i for i in range(-1000, 1000)]
y = [np.sin(i) for i in x]

ax.plot(x, y)
```

limitando, fica:

```python
ig, ax = plt.figure(), plt.axes()

x = [i for i in range(-1000, 1000)]
y = [np.sin(i) for i in x]

plt.xlim(-10, 10)
plt.ylim(-1.5, 1.5)

ax.plot(x, y)
```

se desejar mostrar os valores em ordem decrescente, basta invester os valores: `xlim(max, lim)` e `ylim(max, lim)`:

```python
ig, ax = plt.figure(), plt.axes()

x = [i for i in range(-1000, 1000)]
y = [np.sin(i) for i in x]

plt.xlim(10, -10)
plt.ylim(1.5, -1.5)

ax.plot(x, y)
```

```python
fig, ax = plt.figure(), plt.axes()

a = [i for i in range(-1000, 1000)]
b = [np.sin(i) for i in x]

plt.axis([-10, 10, -1.2, 1.2])

ax.plot(a, b)
```

passando a string `tight` para `plt.axis()`, o gráfico aparece justificado:

```python
fig, ax = plt.figure(), plt.axes()

a = [i for i in range(0, 10)]

plt.axis('tight')

ax.plot(a, np.sin(a))
```

#### textos no gráfico
---

usando a função `.text()` é possível mostrar um texto dentro do gráfico. É, para isso, necessário passar as posições x e y, a mensagem, e outras informações opcionais como cor, tamanho, estilo, alinhamento, etc.:

```python
fig, ax = plt.figure(), plt.axes()

a = [i for i in range(0, 11)]

ax.text(4, -.75, 'f(4)=-0.75', ha='center', va='center')
ax.plot(a, np.sin(a))
```

o parâmetro `ha=` localiza a mensagem de acordo com a coordenada x passada, e deve receber `left`, `center` e `right`. Também pode passar o parâmetro `va=`, que localiza a mensagem de acordo com a coordenada y passada, e deve receber `top`, `center` e `bottom.`

é possível escrever esses textos usando os pixels do gráfico caso não queira depender dos dados armazenados. Para isso, deve-se passar para o parâmetro `transform=` um dos argumentos: `ax.transAxes`, se desejar associar aos eixos; `ax.transFigure`, se desejar associar à janela; ou o `ax.transData`, que é o padrão, não precisando passar.

```python
fig, ax = plt.figure(), plt.axes()

a = [i for i in range(0, 11)]

# neste caso, os valores de x e y passados será referente os pixels.
ax.text(0.4, 0.14, 'TransAxes', ha='center', transform=ax.transAxes)
ax.text(0.51, 0.13, 'TransFigure', ha='center', transform=fig.transFigure)
ax.text(8, 1, 'TransData', ha='center', transform=ax.transData)
ax.plot(a, np.sin(a))
```

logo, por exemplo, se mudar a escala da janela, `TransAxes` e `Transfigure` não mudam de lugar, mas `TransData` muda:

```python
fig, ax = plt.figure(), plt.axes()

a = [i for i in range(0, 11)]

ax.set_xlim(0, 2)
ax.set_ylim(-6, 6)

# neste caso, os valores de x e y passados será referente os pixels.
ax.text(0.4, 0.14, 'TransAxes', ha='center', transform=ax.transAxes)
ax.text(0.51, 0.13, 'TransFigure', ha='center', transform=fig.transFigure)
ax.text(8, 1, 'TransData', ha='center', transform=ax.transData)
ax.plot(a, np.sin(a))
```

#### anotações e setas
---

a função `annotate()` também recebe um texto e uma coordenada (x, y) no parâmetro `xy=` para mostrar textos no gráfico:

```python
fig, ax = plt.figure(), plt.axes()

a = [i for i in range(0, 11)]

ax.plot(a, np.sin(a))

ax.annotate('máximo local', xy=(8, 1))
```

no entanto, esta função também pode desenhar setas no gráfico, para isso, é necessário localizar o texto através do parâmetro `xytext=`, pois `xy=` passará a servir para localizar a seta, e a seta será feita através do parâmetro `arrowprops=` que deve receber um dicionário contendo as especificações da seta.

```python
fig, ax = plt.figure(), plt.axes()

a = [i for i in range(0, 11)]

ax.plot(a, np.sin(a))

ax.annotate('máximo local', xy=(8, 1), xytext=(5, 0.75), arrowprops={'facecolor': 'black', 'shrink': 0.05})
```

`facecolor` recebe uma cor para o preenchimento da seta, `shrink` recebe um valor entre 0 e 1 e modifica o tamanho do corpo da seta.

`.annotate()` recebe todos os parâmetros já visto.

#### ticks
---

os chamados *ticks*, ou marcações, são os valores que aparecem nos eixos x e y e podem ser manipulados através do objeto `axes()`.

Uma das principais manipulações é a mudança de escala quando os  valores são muito pequenos ou muito grandes, pode ser feito passando os parâmetros `xscale=`, que modifica o eixo x, e o `yscale=`, que modifica o eixo y.

```python
ax = plt.axes(xscale='log', yscale='log')
ax.grid()
ax.set_xlim(10**-3, 10**3)
ax.set_ylim(10**-3, 10**3)
```

os objetos `formatter` e `locator` servem para manipular a localização e a formatação desses valores. Estes, por exemplo podem ser usados para esconder esses valores passando os argumentos `plt.NullLocator()` para o método `.set_major_locator()`, que serve para esconder os traços que localizam os valores e os valores, e `plt.NullFormatter()` para o método `set_major_formatter()`, que serve para esconder apenas os valores, mas os traços permanecem. O eixo que em que estes método serão aplicados serão escolhidos passando `.xaxis` e/ou `.yaxis` antes dos métodos.

```python
ax = plt.axes()
ax.plot(np.random.rand(50))

ax.xaxis.set_major_locator(plt.NullLocator())
ax.yaxis.set_major_formatter(plt.NullFormatter())
```

usando a função `plt.MaxNLocator(<número>)`, é possível passar o número máximo de ticks por eixo:

```python
fig, ax = plt.subplots(nrows=4, ncols=4, sharex=True, sharey=True)
```

agora, observe a diferença quando passado esta função:

```python
fig, ax = plt.subplots(nrows=4, ncols=4, sharex=True, sharey=True)

for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(3))
    axi.yaxis.set_major_locator(plt.MaxNLocator(3))
```

a função `plt.MultipleLocator()` recebe um valor, o que fará com que todos os valores apresentados sejam multiplos deste valor:

```python
ax = plt.axes()

ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/4))
ax.set_xlim(0, 10)
```

neste exemplo, usando os valores de $π$, vê-se que as casas decimais podem ser um problema. Neste caso, não há uma direta de fazer aparecer o símbolo de python, mas a função `plt.FuncFormatter()` consegue receber uma função criada pelo usuário que permita modificar esses ticks na forma desejada:

```python
# Plot a sine and cosine curve
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.plot(x, np.cos(x), lw=3, label='Cosine')

# Set up grid, legend, and limits
ax.grid(True)
ax.legend(frameon=False)
ax.axis('equal')
ax.set_xlim(0, 10);

def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
```
