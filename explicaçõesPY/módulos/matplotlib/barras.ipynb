## matplotlib.pyplot: bar
---

````python
import matplotlib.pyplot as plt
````

quando for mostrar o gráfico, ao invés de usar `plt.plot()`, pode-se usar `plt.bar()` para mostrar as informações em barras:

````python
x = [i for i in range(-5, 6)]
y = [i**2 for i in range(-5, 6)]

plt.bar(x, y)
plt.show()
````

inclusive, várias formas de gráficos podem ser misturados:

````python
x = [x for x in range(-5, 6)]
y1 = [i**2 for i in x]
y2 = [i**2 + 5*i for i in x]

plt.bar(x, y1)
plt.plot(x, y2)
plt.show()
````

às vezes, plotar duas informações em um só gráfico usando o tipo barra pode ser problemático, pois uma informações pode cobrir outra:

````python
x = [x for x in range(-5, 6)]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.bar(x, y1)
plt.bar(x, y2)
plt.show()
````

para evitar isso é necessário usar o numpy

````python
import numpy as np
````

e seguir o código:

````python
x = [x for x in range(-5, 6)]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

x_indexes = np.arange(len(x))
width = 0.25

plt.bar(x_indexes - width, y1, width=width)
plt.bar(x_indexes, y2, width=width)
plt.show()
````

observe que foi acrescentado as linhas
```
x_indexes = np.arange(len(x))
width = 0.25
```
onde `x_indexes` serve para criar uma lista com índices de 0 até o último elemento da lista original e esta variável deve usada no lugar de `x` quando for plotar o gráfico.
`width` servirá para deslocar `x_indexes` na visualização do gráfico, por isso que, neste exemplo, só é usado com os valores quadrados pois, com estes se deslocando, já é suficiente para visualizar todas as plotagens.

observe que os valores da abscissas estão errados, pois os valores que queremos observar vai de -5 a 5, porém, está apresentando de 0 a 10. Isto é devido estarmos usando `x_indexes` ao invés de `x`.

Para corrigir este problema é necessário acrescentar `plt.xticks(ticks=x_indexes, labels=x)`

````python
x = [x for x in range(-5, 6)]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

x_indexes = np.arange(len(x))
width = 0.25

plt.bar(x_indexes - width, y1, width=width)
plt.bar(x_indexes, y2, width=width)
plt.xticks(ticks=x_indexes, labels=x)
plt.show()
````

para colocar as barras na horizotal, basta substituir `plt.bar()` por `plt.barh()`

````python
x = [i for i in range(-5, 6)]
y = [i**2 for i in x]

plt.barh(x, y)
plt.show()
````
